```python
class divisible:
    def __init__(self,n,start,end):
        self.n=n
        self.start=start
        self.end=end
    def calculate(self):
        for i in range(self.start,self.end+1):
            if(i%7==0):
                print(i)
            
m=divisible(7,0,100)
m.calculate()
```

```python
def binary_search(arr, x):
    low = 0
    high = len(arr) - 1
    mid = 0
    while low <= high:
        mid = (high + low) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
arr = [ 2, 3, 4, 10, 40 ]
x = 10
result = binary_search(arr, x)
if result != -1:
    print("Element is present at index", str(result))
else:
    print("Element is not present in array")
```
